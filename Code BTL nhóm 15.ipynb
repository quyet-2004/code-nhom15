```python
import random

def introduce_typo(word):
    """Introduce random typos into a word."""
    if len(word) <= 1:
        return word

    typo_type = random.choice(["swap", "remove", "duplicate", "replace"])
    index = random.randint(0, len(word) - 1)

    if typo_type == "swap" and index < len(word) - 1:
        # Swap two adjacent characters
        return word[:index] + word[index + 1] + word[index] + word[index + 2:]
    elif typo_type == "remove":
        # Remove a character
        return word[:index] + word[index + 1:]
    elif typo_type == "duplicate":
        # Duplicate a character
        return word[:index] + word[index] + word[index] + word[index + 1:]
    elif typo_type == "replace":
        # Replace a character with a random letter
        random_char = random.choice("abcdefghijklmnopqrstuvwxyz")
        return word[:index] + random_char + word[index + 1:]
    return word

def generate_typo_dataset(sentences):
    """Generate a dataset with typos."""
    typo_sentences = []
    for sentence in sentences:
        words = sentence.split()
        typo_sentence = " ".join([introduce_typo(word) for word in words])
        typo_sentences.append((typo_sentence, sentence))  # (noisy_sentence, clean_sentence)
    return typo_sentences

# Example usage
sentences = ["Hôm nay tôi đi học", "Tôi thích lập trình Python"]
typo_dataset = generate_typo_dataset(sentences)
print(typo_dataset)
```

```python
import nltk
from nltk.corpus import words

# Ensure the 'words' corpus is downloaded
try:
    nltk.data.find('corpora/words.zip')  # Check if the corpus already exists
except LookupError:
    nltk.download('words')  # Download if not found

# Load the English words corpus
english_words = set(words.words())

def detect_spelling_errors(tokens):
    """Detect misspelled words in a list of tokens."""
    errors = [token for token in tokens if token.lower() not in english_words]
    return errors

# Example usage
tokens = ["hôm", "nay", "toi", "di", "hoc"]
errors = detect_spelling_errors(tokens)
print("Misspelled words:", errors)
```

```python
import re
from difflib import get_close_matches

# Thay vì import, định nghĩa hàm trực tiếp ở đây
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = text.split()
    return tokens

def detect_spelling_errors(tokens):
    return [token for token in tokens if token not in vocabulary]

def correct_spelling(word, vocabulary):
    matches = get_close_matches(word, vocabulary, n=1, cutoff=0.7)
    return matches[0] if matches else word

# Example vocabulary (có thể mở rộng từ vựng này)
vocabulary = set(["hôm", "nay", "tôi", "đi", "học", "nhưng", "quên", "sách"])

def spelling_correction_pipeline(text):
    """Pipeline xử lý phát hiện và sửa lỗi chính tả."""
    tokens = preprocess_text(text)
    errors = detect_spelling_errors(tokens)
    corrected_tokens = [
        correct_spelling(token, vocabulary) if token in errors else token for token in tokens
    ]
    corrected_text = " ".join(corrected_tokens)
    return corrected_text

# Ví dụ sử dụng
if __name__ == "__main__":
    input_text = "Hôm nay toi di hoc nhung quên sach"
    corrected_text = spelling_correction_pipeline(input_text)
    print("Corrected text:", corrected_text)
```

```python
import tkinter as tk
from tkinter import messagebox
import re
from difflib import get_close_matches
import unicodedata

# ======== Bỏ dấu tiếng Việt ========
def remove_vietnamese_accents(text):
    return ''.join(
        c for c in unicodedata.normalize('NFD', text)
        if unicodedata.category(c) != 'Mn'
    )

# ======== Tạo từ điển có dấu và không dấu ========
vocab_with_diacritics = ["hôm", "nay", "tôi", "đi", "học", "nhưng", "quên", "sách"]
vocab_no_diacritics = [remove_vietnamese_accents(w) for w in vocab_with_diacritics]
vocab_dict = dict(zip(vocab_no_diacritics, vocab_with_diacritics))
vocab_set = set(vocab_no_diacritics)

# ======== Tiền xử lý ========
def preprocess_text(text):
    text = text.lower()
    text = remove_vietnamese_accents(text)
    text = re.sub(r'[^\w\s]', '', text)
    return text.split()

# ======== Phát hiện và sửa lỗi ========
def detect_spelling_errors(tokens, vocabulary):
    return [token for token in tokens if token not in vocabulary]

def correct_spelling(word, vocabulary):
    matches = get_close_matches(word, vocabulary, n=1, cutoff=0.7)
    return matches[0] if matches else word

def spelling_correction_pipeline(text):
    tokens = preprocess_text(text)
    errors = detect_spelling_errors(tokens, vocab_set)
    corrected_tokens = [
        correct_spelling(token, vocab_set) if token in errors else token
        for token in tokens
    ]
    # Khôi phục dấu từ mapping
    corrected_tokens_with_diacritics = [
        vocab_dict.get(token, token) for token in corrected_tokens
    ]
    return " ".join(corrected_tokens_with_diacritics)

# ======== GUI ========
def correct_spelling_gui():
    input_text = input_textbox.get("1.0", tk.END).strip()
    if not input_text:
        messagebox.showerror("Lỗi", "Vui lòng nhập nội dung vào hộp văn bản.")
        return

    corrected_text = spelling_correction_pipeline(input_text)
    output_textbox.delete("1.0", tk.END)
    output_textbox.insert(tk.END, corrected_text)

# ======== Cửa sổ chính ========
root = tk.Tk()
root.title("Công cụ sửa lỗi chính tả")

input_label = tk.Label(root, text="Nhập văn bản:")
input_label.pack(pady=5)

input_textbox = tk.Text(root, height=10, width=50, wrap=tk.WORD)
input_textbox.pack(padx=10, pady=5)

correct_button = tk.Button(root, text="Sửa lỗi chính tả", command=correct_spelling_gui)
correct_button.pack(pady=10)

output_label = tk.Label(root, text="Văn bản sau khi sửa:")
output_label.pack(pady=5)

output_textbox = tk.Text(root, height=10, width=50, wrap=tk.WORD)
output_textbox.pack(padx=10, pady=5)

root.mainloop()
```
